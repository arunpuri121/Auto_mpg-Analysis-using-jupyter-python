```python
# importing packages
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# importing warnings
import warnings
warnings.filterwarnings('ignore')
pd.set_option("display.max_columns", None)
pd.set_option("display.max_rows", None)
```

```python
# Importing Data Set
df = pd.read_csv('auto-mpg.csv.xls')
```

```python
df.isnull().sum()
```

```python
df['hp'].value_counts()
```

```python
df['hp'] = df['hp'].replace('?', np.nan)
df['hp'] = df['hp'].astype(float)
```

```python
df['hp'].dtypes
```

```python
df.describe()
```

```python
df['hp'] =df['hp'].replace(np.nan, df['hp'].median())
```

```python
df.describe()
```

```python
df['origin']=df['origin'].replace({1:'America', 2:'China', 3:'India'})
```

```python
df.sample(10)
```

```python
df.hist(figsize=(15,10))
plt.show()
```

```python
df.skew(numeric_only=True)
```

```python
sns.boxplot(x='mpg', data=df)
plt.show()
```

```python
sns.boxplot(x='cyl',data= df)
plt.show()
```

```python
import plotly.express as px
```

```python
for column in df:
    fig=px.histogram(df,x=column, nbins=20,)
    fig.show()
```

```python
for column in df:
    fig=px.box(df,x=column)
    fig.show()
```

```python
sns.pairplot(df)
plt.show()
```

```python
sns.violinplot(x="origin", y="mpg",data=df)
plt.show()
```

```python
corr_matrix=df.corr(numeric_only=True)
corr_matrix
```

```python
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm")
plt.show()
```

```python
sns.kdeplot(x="wt", y="mpg", data=df)
plt.show()
```

```python
sns.boxenplot(x="origin", y="mpg", data=df)
plt.show()
```

```python
g = sns.FacetGrid(col='origin', row='yr',data=df)
g.map(sns.scatterplot, 'cyl', 'acc')
g.fig.subplots_adjust(top=0.9)
g.fig.suptitle('FacetGrid Plot')
plt.show()
```

```python
sns.pairplot(data=df, hue="mpg", diag_kind="kde", palette="husl")
plt.show()
```
